```
dataset_source = pd.read_csv("./drive/My Drive/Agora-Datasets/my_dataset.csv")
is_trust_worthy = dataset_source['confidence'] >= 0.9
rows_that_we_trust = dataset_source[is_trust_worthy]
trust_count = dict()
for label_name in label_to_cat.keys():
  trusty_bool = rows_that_we_trust['label_display_name'] == label_name
  new_rows = rows_that_we_trust[trusty_bool]
  trust_worthy_ids =set(new_rows['image_id'])
  if label_to_cat[label_name]  not in trust_count:
     trust_count[label_to_cat[label_name]] = 0
  trust_count[label_to_cat[label_name]] += len(trust_worthy_ids)
for k, v in trust_count.items():
  print("Trust Worthy {}s are {}".format(k, v))
```
