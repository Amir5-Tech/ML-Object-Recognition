```
from google_images_download.google_images_download import googleimagesdownload
response = googleimagesdownload()
def download_image(query):
  arguments = {
      'keywords': query,
      'limit':100,
      'print_urls': True,
  }
  try:
    response.download(arguments)
  except FileNotFoundError:
    arguments = {"keywords": query, 
                 "format": "jpg", 
                 "limit":100, 
                 "print_urls":True,  
                 "size": "medium",
                } 
        # Providing arguments for the searched query 
    try: 
            # Downloading the photos based 
            # on the given arguments 
      response.download(arguments)  
    except: 
      pass
```

```
from google.colab import files
from os import listdir
from os.path import isfile, join
import shutil
```

```
parent_folders_to_move = {'keychain natural setting': 'Key', 'soda can': 'Can', 'keychain keys': 'Key', 'soda cans': 'Can',
                         'opened soda cans': 'Can', "keys in hand": 'Key', "beverage can on table": "Can"}
```

```
root_f_path = "./downloads"
root_d_path = "./drive/My Drive/Agora-Datasets/open_images"
for f_folder, to_folder in parent_folders_to_move.items():
  f_fp = join(root_f_path, f_folder)
  for file in listdir(f_fp):
    shutil.move(join(f_fp, file), join(root_d_path, to_folder, file))  
    
```
